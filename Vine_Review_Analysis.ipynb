```python
import pandas as pd
```

```python
vine_df = pd.read_csv('https://raw.githubusercontent.com/LJD0/Amazon_Vine_Analysis/main/Resources/vine_table.csv')
vine_df
```

```python
# Filter reviews that have more than 20 total votes to find more helpful reviews

vine_df_over_20 = vine_df[vine_df['total_votes']>20]
vine_df_over_20
```

```python
# Find the most helpful reviews

helpful_df = vine_df_over_20[vine_df_over_20['helpful_votes']/vine_df_over_20['total_votes']>=0.5]
helpful_df
```

```python
# find the paid reviews (vine==y)

vine_y = helpful_df[helpful_df['vine']=='Y']
vine_y
```

```python
# find the un-paid reviews (vine==n)

vine_n = helpful_df[helpful_df['vine']=='N']
vine_n
```

```python
vine_y_len = len(vine_y)
vine_y_five_len = len(vine_y[vine_y['star_rating']==5])
vine_n_len = len(vine_n)
vine_n_five_len = len(vine_n[vine_n['star_rating']==5])
```

```python
paid_summary = f'''
  The total number of paid reviews (vine==Y) is: {vine_y_len}
  The total number of paid 5 star reviews is: {vine_y_five_len}
  The percent of paid reviews that are five stars is: {100*round(vine_y_five_len/vine_y_len,4)}%
  '''
unpaid_summary = f'''
  The total number of unpaid reviews (vine==N) is: {vine_n_len}
  The total number of unpaid 5 star reviews is: {vine_n_five_len}
  The percent of unpaid reviews that are five stars is: {100*round(vine_n_five_len/vine_n_len,4)}%
  '''
```

```python
print(paid_summary)
```

```python
print(unpaid_summary)
```
